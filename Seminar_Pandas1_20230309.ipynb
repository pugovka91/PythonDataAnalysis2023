Заменим F/M на 0/1 в столбце Пол

```python
df['CODE_GENDER_BINARY'] = np.where(df['CODE_GENDER']=='F', 1, 0)
df['CODE_GENDER_BINARY'].value_counts()
```

```python
df['CODE_GENDER_BINARY1'] = df['CODE_GENDER'].apply(lambda x: 1 if x=='F' else 0)
df['CODE_GENDER_BINARY1'].value_counts()
```

```python
gender_dict = {
    'F':1,
    'M':0
}

df['CODE_GENDER_BINARY2'] = df['CODE_GENDER'].map(gender_dict)
df['CODE_GENDER_BINARY2'].value_counts()
```

Составить столбец на основе наличия автомобиля и дохода: 
* если доход выше среднего, и есть авто
* доход выше среднего и нет авто
* есть авто, доход ниже среднего
* нет авто и доход ниже среднего

```python
df.head(2)
```

```python
def bin_auto_income_func(x, mean_income):
    if x['amt_income_total']>mean_income and x['flag_own_car']=='Y':
        res = 'income_car'
    elif x['amt_income_total']>mean_income:
        res = 'income'
    elif x['flag_own_car'] == 'Y':
        res = 'car'
    else:
        res = 'no_income_car'
    return res
```

```python
df['bin_auto_income'] = df.apply(lambda x: bin_auto_income_func(x, df['amt_income_total'].mean()), axis = 1)
```

```python
df['bin_auto_income'].value_counts()
```

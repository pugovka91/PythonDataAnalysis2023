### Data

```python
df = pd.read_csv(application_train_data_path)
```

```python
df.head(2)
```

```python
df_desc = pd.read_csv(columns_desc_data_path, encoding = 'windows-1251')
df_desc
```

```python
need_cols = [
    'sk_id_curr',
    'target',
    'name_contract_type',
    'code_gender',
    'flag_own_car',
    'flag_own_realty',
    'cnt_children',
    'amt_income_total',
    'amt_credit',
    'amt_annuity',
    'amt_goods_price',
    'name_type_suite',
    'name_income_type',
    'name_education_type',
    'name_family_status',
    'name_housing_type',
    'region_population_relative',
    'days_birth',
    'days_employed',
    'days_registration',
    'days_id_publish',
    'own_car_age',
    'flag_mobil',
    'flag_emp_phone',
    'flag_work_phone',
    'flag_cont_mobile',
    'flag_phone',
    'flag_email',
    'occupation_type',
    'cnt_fam_members',
    'amt_req_credit_bureau_hour',
    'amt_req_credit_bureau_day',
    'amt_req_credit_bureau_week',
    'amt_req_credit_bureau_mon',
    'amt_req_credit_bureau_qrt',
    'amt_req_credit_bureau_year'
]
```

```python
#отберем часть столбцов для дальнейшей работы
df.columns = [x.lower() for x in df.columns]
df = df[need_cols]
```

```python
df.shape
```

```python
#небольшой препроцессинг с прошлого занятия
cols_to_preprocess = ['days_birth', 'days_employed','days_registration']

for x in cols_to_preprocess:
    df[f'{x}_years'] = np.abs(df[x])/365.25
```

```python
df['age_bins'] = pd.cut(
    df['days_birth_years'], 
    bins = [-np.inf, 23, 25, 30, 35, 40, 45, 50, np.inf],
    labels = ['0.0-23', '1.23-25', '2.25-30', '3.30-35', '4.35-40', '5.40-45', '6.45-50', '7.50+']
)
```

#### Несколько графиков на одной картинке

```python
plt.rcParams["figure.figsize"] = (6, 6)
plt.rcParams['figure.constrained_layout.use'] = True

# constrained_layout automatically adjusts subplots and decorations like legends and colorbars so that they fit in 
# the figure window while still preserving, as best they can, the logical layout requested by the user.
```

```python
fig, ax = plt.subplots(2, 1) #nrows, ncols

ax[0].bar(dr['age_bins'], dr['amt_income_total'])
ax[0].set_xlabel('age_bins')
ax[0].set_ylabel('amt_income_total')

def silly_func(t):
    res = np.exp(t)*np.sin(t)
    return res

x_array = np.arange(0, 10)

ax[1].scatter(x_array, silly_func(x_array), color  = 'green')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')

fig.show()
```

```python
plt.rcParams["figure.figsize"] = (12, 6)
fig, ax_dict = plt.subplot_mosaic(
    [
        ['bar', 'scatter'],
        ['hist1', 'hist2'],
    ],
)
ax_dict['bar'].bar(dr['age_bins'], dr['amt_income_total'])
ax_dict['scatter'].plot([1, 2, 3])
ax_dict['hist1'].hist(df['amt_req_credit_bureau_year'])
ax_dict['hist2'].hist(df['amt_req_credit_bureau_mon'])

fig.suptitle('Subplots mosaic') 
fig.show()
```

#### Две оси y на одном графике

```python
plt.rcParams["figure.figsize"] = (6, 4)
fig, ax1 = plt.subplots()


ax2 = ax1.twinx()
ax1.plot(dr['age_bins'], dr['amt_income_total'], color = 'g')
ax2.plot(dr['age_bins'], dr['amt_credit'], color = 'b')

ax1.set_xlabel('age_bins')
ax1.set_ylabel('amt_income_total', color='g')
ax2.set_ylabel('amt_credit', color='b')
fig.show()
```

```python
plt.rcParams["figure.figsize"] = (7, 4)
fig, ax1 = plt.subplots()


ax2 = ax1.twinx()
x = np.arange(len(dr['age_bins']))

ax1.bar(x-0.15, dr['amt_income_total'], color = 'g', width = 0.3)
ax2.bar(x+0.15, dr['amt_credit'], color = 'b', width = 0.3)

ax1.set_xticklabels(dr['age_bins'].astype('str').values)
ax1.set_xticks(x)

ax1.set_xlabel('age_bins')
ax1.set_ylabel('amt_income_total', color='g')
ax2.set_ylabel('amt_credit', color='b')
fig.show()
```
